```python
movies = movies[['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'crew']]
movies.shape
```

```python
def convert(item):
    l = []
    item = eval(item)
    for i in item:
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head(1)
```

```python
def getDirector(item):
    l = []
    item = eval(item)
    for i in item:
        if i["job"] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(getDirector)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movies.head(1)
```

```python
def removeComma(arr):
    l = []
    for i in arr:
        l.append(i.replace(',', ''))
    return l
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split(' '))
movies['overview'] = movies['overview'].apply(removeComma)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew'] 
```

```python
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x))
```

```python
pdf = movies[['id', 'title', 'tags']]
```

```python
pdf.head()
```

```python
pdf['tags'] = pdf['tags'].apply(lambda x: x.lower())
```

```python
pdf.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemmer(text):
    l = []
    for i in text.split(' '):
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
pdf['tags'] = pdf['tags'].apply(stemmer)
```

```python
pdf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(pdf['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    index = pdf[pdf['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(pdf.iloc[i[0]].title)
```

```python
recommend('Charlie and the Chocolate Factory')
```
